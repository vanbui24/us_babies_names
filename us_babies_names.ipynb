```python
#In CA: what were the most popular male and female babies names in each year

# 1 sort by CA (Slice out rows for CA)
# 2 grup by year and gender
# 3 for each group compute popular for each group

# 1 sort by CA (Slice out rows for CA)
ca_babies = states_babies.loc[states_babies['State'] == "CA", : ]
```

```python
ca_babies
```

```python
def popular(s):
    """Recieves s, a Pandas Series, contaning baby names in order of highest count to lowest count.
    Returng the most popular name in s."""
    return s.iloc[0]
```

```python
#group by year and gender
#group by method
#group by year and gender

ca_grouped = ca_babies.sort_values('Count', ascending = False).groupby(['Year', 'Gender']).agg(popular)
```
